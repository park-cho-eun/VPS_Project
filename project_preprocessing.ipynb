```python
import pandas as pd
import re
from nltk.tokenize import sent_tokenize
```

```python
data1=pd.read_excel('C:/Users/ssb70/OneDrive/바탕 화면/간호법/data1.xlsx')
data2=pd.read_excel('C:/Users/ssb70/OneDrive/바탕 화면/간호법/data2.xlsx')
```

```python
data1.columns=data1.iloc[0]
data1=data1[1:]
data2.columns=data2.iloc[0]
data2=data2[1:]

data=pd.concat([data1,data2],axis=0,ignore_index=True)
```

```python
class Preprocessing():
    def __init__(self):
        self.data= None

    def forward(self,data):
        data['내용_전처리(특수문자)']=None
        data['내용'] = data['내용'].astype(str)
        # 마침표를 제외한 다른 특수 문자 제거
        for i, text in enumerate(data['내용']):
            data.loc[i, '내용_전처리(특수문자)'] = re.sub(r'[^\w\s.]', '', text)

        data['slicing']=None
        # sent_tokenize를 이용해 문장 토큰화 ('.'이 있으므로 개빠름)
        for i, text in enumerate(data['내용_전처리(특수문자)']):
            texts=sent_tokenize(text)
            data.at[i,'slicing']=texts
        # 기사의 문장 수가 단 1줄인 기사는 제외
        data['length'] = data['slicing'].apply(len)
        data = data[data['length'] > 1].reset_index(drop=True)
        data.drop(columns=['length'], inplace=True)
        
        # embedding 열도 생성하기
        data['embedding']=None
        
        return(data)
        

```

```python
pre=Preprocessing()
data=pre.forward(data)
```
